```python
# 划分特征
x = df_train.iloc[:,:-1]
y = df_train['target']
x.sample(3)
```

```python
# 特征相关性分析

pearson = x.corr(method="pearson") #计算皮尔逊相关系数

for i in range(0,38):
    pearson.iloc[:i+1,i] = np.nan   #取对角矩阵

display( pearson )

#绘制热力图
plt.figure( figsize=(20,9), dpi=150 )
sns.heatmap( pearson, cmap='rainbow' )
plt.show()
```

```python
# 递归赛选特征
from sklearn.feature_selection import RFECV
from sklearn.ensemble import RandomForestRegressor

model_rfc = RandomForestRegressor( random_state=111 ) # 使用随机深林为基模型
rfc_cv = RFECV( estimator=model_rfc, min_features_to_select=20, cv=10, n_jobs=6 ) # 使用10折交叉验证最少选取20个特征
x1 = rfc_cv.fit_transform( x,y )

print( '是否被选中:\n',rfc_cv.support_ )
print( '\n数据尺度 : ',x1.shape )
```

```python
x1 = x.loc[:,rfc_cv.support_]
x1
```

```python
# 使用lasso回归进行再选择
from sklearn.linear_model import LassoCV

lasso = LassoCV( cv=10, alphas=np.logspace(-6, 6, 13), random_state=111 ) #构建lasso回归网格搜索
lasso.fit( x1,y )

print( f'alphas = {lasso.alpha_}' ) #输出alphas的值

#将特征名、系数、系数绝对值生成DataFrame，并排序
lasso_df = pd.DataFrame( {'特征名' : x.columns[rfc_cv.support_] ,
                          '系数' : lasso.coef_,
                          '系数绝对值' : abs(lasso.coef_) } ).sort_values( by='系数绝对值',ascending=False )

display( lasso_df )
```

```python
# 用网格搜索来赛选特征数
from sklearn.model_selection import cross_val_score

threshould = np.linspace( start=lasso_df['系数绝对值'].min(), stop=lasso_df['系数绝对值'].max()-0.001, num=30 ) #生成等差数列
score = []

for i in threshould:
    index = lasso_df[ lasso_df['系数绝对值'] > i ]['特征名'] # 根据系数绝对值赛选特征
    x2 = np.array( df_train[index] ) # 选择数据
    once = cross_val_score( model_rfc, x2, y, cv=10 ) # 使用10折交叉验证赛选，模型为随机深林
    score.append( np.mean(once) )
```

```python
# 可视化
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.figure( figsize=(10,4),dpi=300 )
plt.plot( threshould,score )
plt.xticks(threshould.tolist(),rotation=90)
plt.show()
```

```python
# 选择特征
index = lasso_df[ lasso_df['系数绝对值'] > 0.0634 ]['特征名'] # 赛选‘系数绝对值’大于0.0634的值
index
```

```python
# 划分数据集
from sklearn.model_selection import train_test_split

x = df_train[index] # 选择赛选出来的特征

y = df_train['target']
data_test = df_test

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.3, random_state=111)

x_train.sample(5)
```

```python
# 选择模型
from sklearn.metrics import r2_score

def mae(y_true, y_pred):
    return np.mean( abs(y_true - y_pred) ) # 计算平均绝对值误差

def fit_and_evaluate( model ):
    model.fit( x_train, y_train ) # 建模

    model_pred = model.predict( x_test )
    model_mae = mae( y_test, model_pred )
    r2 = r2_score( y_test, model_pred )

    return model_mae,r2  # 返回平均绝对值误差
```

```python
# 模型选择
scores = [] # 储存平均绝对值误差
scores_r2 = [] # 储存R方

from sklearn.linear_model import LinearRegression
lr = LinearRegression()  # 一般线性回归
lr_mae,r2 = fit_and_evaluate(lr)
scores.append(lr_mae)
scores_r2.append(r2)

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()  # 决策树回归
dt_mae,r2 = fit_and_evaluate(dt)
scores.append(dt_mae)
scores_r2.append(r2)

from sklearn.ensemble import RandomForestRegressor
random_forest = RandomForestRegressor()  #随机深林回归
random_forest_mae,r2 = fit_and_evaluate(random_forest)
scores.append(random_forest_mae)
scores_r2.append(r2)

from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor() # K近邻回归
knn_mae,r2 = fit_and_evaluate(knn)
scores.append(knn_mae)
scores_r2.append(r2)

from xgboost import XGBRegressor
Xgboost = XGBRegressor() # xgboost回归
Xgboost_mae,r2 = fit_and_evaluate(Xgboost)
scores.append(Xgboost_mae)
scores_r2.append(r2)

types = ['LinearRegression', 'DecisionTreeRegressor', 'RandomForestRegressor',
         'KNeighborsRegressor', 'XGBRegressor']

score = pd.DataFrame({'MAE': scores, 'R方' : scores_r2 }, index=types) # 生成DataFrame

ax = score.sort_values( by='MAE', ascending=False ).plot.barh( grid=True,figsize=(8,6) ) # 画出横向柱状图
ax.set_xticks( np.linspace(0,1,10) )
ax.legend(loc='lower right')
plt.show()
```

```python
# 线性回归模型调参
from sklearn.model_selection import GridSearchCV

params = {
    'fit_intercept' : [True,False],
    'normalize' : [True,False]
}

model_lr = LinearRegression( copy_X=True,positive=False ) #线性回归模型

grid_search = GridSearchCV( model_lr, params, cv=10, scoring='neg_mean_absolute_error',
                            n_jobs = 6, refit=True )# 网格搜索
grid_search.fit( x_train, y_train )

grid_search.best_params_
```

```python
# 线性回归模型建模
model_lr = LinearRegression( copy_X=True, fit_intercept=True, normalize=True, positive=False ) #线性回归模型
model_lr.fit(x_train,y_train)
```

```python
print( '调参前:\n',score.loc['LinearRegression'] )
```

```python
# 线性回归模型评价
from sklearn import metrics

print( '模型分数 : ', model_lr.score( x_test,y_test ) )
print( '平均绝对值误差 : ',  metrics.mean_absolute_error( y_test, model_lr.predict( x_test ) ))
print( '平均平方误差 : ',  metrics.mean_squared_error( model_lr.predict(x_test), y_test ) )
print( '中位绝对值误差 : ',  metrics.median_absolute_error( model_lr.predict(x_test), y_test ) )
print( 'R2得分 : ',  metrics.r2_score( model_lr.predict(x_test), y_test ) )
```

```python
# 预测结果可视化
plt.figure( figsize=(9,5),dpi=200 )

sns.kdeplot( y_test, label = '真实值')
sns.kdeplot( model_lr.predict( x_test ), label = '预测值')

plt.xlabel('Energy Star Score')
plt.ylabel('工业蒸汽量')
plt.legend()
plt.title('Test Values and Predictions');
```
